# VIDEO: String Interpolation


```python
# review on string formatting

num = 45.5
sng = 'cellar doors'

print( 'I would like to have ' + str(num) + ' ' + sng )
```

```python
# String Interpolation
# Better way for Maintainability

# i (integer), g (number), s (string)
print( 'I would like to have %g %s.' %(num,sng) ) 
# f (10.3 display 3 decimal places rounded with 10 spaces occupied)
print( 'I would like to have %10.3f %s.' %(num,sng) )
```

```python
print( 'I would like to have %g %s.'%(num**2/18.5,sng) )
```

```python
# f-strings
# Better way for Maintainability

print(f'I would like to have {num} {sng}.')
```

```python
# Formatted string
print('I would like to have {n} {s}.'.format(n=num,s=sng))
```

```python
# Exercise
# print out ech letter of the alphabet with its ordinal position
# a is the 1st letter ... z is the 26th letter

import string
letters = string.ascii_lowercase # 'abcd...z'

for i in range(len(letters)):
  if i==0 or i==20:
    ordind = 'st'
  elif i==1 or i==21:
    ordind = 'nd'
  elif i==2 or i==22:
    ordind = 'rd'
  else:
    ordind = 'th'

  # print('%s is the %g%s letter of the alphabet.' %(letters[i],i+1,ordind))
  print(f'{letters[i]} is the {i+1}{ordind} letter of the alphabet.')
```

# VIDEO: Subplot geometry

```python
import matplotlib.pyplot as plt
import numpy as np

# organize curves into a 2x2 subplot figure
# 3rd parameter is the position of the subplot
plt.subplot(2,2,1)
plt.plot(np.random.randn(10))

plt.subplot(2,2,4)
plt.plot(np.random.randn(10))

plt.show()
```

```python
x = np.arange(10)

# Create a 1x3 subplot grid
fig,ax = plt.subplots(1,3,figsize=(15,3)) # (w,h)

ax[0].plot(x,x**2,'b') # y = x^2
ax[0].plot(x,9*x,'k:') # y = 9x
ax[1].plot(x,np.sqrt(x),'r') # y = x^(1/2)
ax[2].plot(x,x) # y = x

plt.show()
```

```python
fig,ax = plt.subplots(2,2)
print(f'Dimension is now: {ax.shape}')
ax[0,0].plot(np.random.randn(4,4), 's:')
ax[0,1].plot(np.random.randn(4,4))
ax[1,0].plot(np.random.randn(4,4), 'k')
ax[1,1].plot(np.random.randn(4,4))

plt.tight_layout() # minimize overlap between different axes (plotting area)
plt.show()
```

# VIDEO: Making the graphs look nicer

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3,3,101)

plt.plot(x,x,label='y=x')
plt.plot(x,x**2,label='y**2')
plt.plot(x,x**3,label='y**3')

plt.legend()
# SMTH NEW: create label in x,y axis, and a title on top of the plot
plt.xlabel('X')
plt.ylabel('y=f(x)')
plt.title('A really awesome plot')

# SMTH NEW: define x,y axis limits
plt.xlim([-3,3]) # xlim[x[0], x[-1]]
plt.ylim([-10,10])

# SMTH NEW sets ratio of x-axis vs. y-axis distance
plt.gca().set_aspect(0.3) #'auto' by default
# plt.axis('square') # square, equal

# SMTH NEW: add grids to the plot
plt.grid() 
plt.show()
```

```python
# Same mechanism applies to the subplots
x = np.linspace(-3,3,101)

fix,ax = plt.subplots()

ax.plot(x,x,label='y=x')
ax.plot(x,x**2,label='y**2')
ax.plot(x,x**3,label='y**3',color=[.8,.1,.7])

ax.legend()
ax.set_xlabel('X')
ax.set_ylabel('y=f(x)')
ax.set_title('A really awesome plot')

ax.set_xlim([-3,3])
ax.set_ylim([-10,10])
ax.set_aspect('equal')
ax.set_aspect(1/ax.get_data_ratio())

ax.grid()

plt.show()
```

# VIDEO: Seaborn


```python
# Library for plotting cleaner numpy arrays & pandas dataframes
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

n = 200
D = np.zeros((n,2))
D[:,0] = np.linspace(0,10,n) + np.random.randn(n)
D[:,1] = D[:,0]**2 + np.random.randn(n)*10

# Joint Plot
# Get marginal distribution of x & y variables
# y is right-skewed due to squared terms 
sns.jointplot(x=D[:,0],y=D[:,1])
plt.show()
```

```python
# Works well on pandas too
import pandas as pd
df = pd.DataFrame(data=D,columns=['var1','var2'])
# uses kde method to estimate probability density function
sns.jointplot(x=df.columns[0],y=df.columns[1],data=df,kind='kde',color='purple')
plt.show()
```

```python
# Create a standard scatter plot with dynamic colors
x = np.linspace(-1,1,n)
y1 = x**2
y2 = np.sin(3*x)
y3 = np.exp(-10*x**2)

# hue determines variable intensity, pallete determines color scheme
sns.scatterplot(x=y1,y=y2,hue=y3,palette='Spectral')
plt.show()
```

```python
# Exercise: make a regression plot in seaborn
sns.regplot(x=df.columns[0],y=df.columns[1],data=df,color='skyblue')
plt.title(f'regression of {df.columns[0]} on {df.columns[1]}')

plt.show()
```

# VIDEO: images

```python
import numpy as np
import matplotlib.pyplot as plt

m = 3
n = 5

M = np.random.randint(10,size=(m,n))
print(M)

# image-show matches each matrix element onto color
plt.imshow(M)
plt.colorbar()

# plot styled text on each matrix element
# be careful that y-axis -> rows, x-axis -> columns
for i in range(m):
  for j in range(n):
    plt.text(j,i,str(M[i,j]),
             horizontalalignment='center',fontsize=20)

plt.show()
```

## Exercise: Visualize the Hilbert matrix using Seaborn

$$H_{i,j} = \frac{1}{i+j-1}$$



```python
# Exercise: Visualize 10x10 Hilbert matrix and show using seaborn heatmap
import seaborn as sns

n = 10

H = np.zeros((n,n))

for i in range(n): 
  for j in range(n):
    H[i,j] = 1 / ((i+1)+(j+1)-1) #note python starts counting at 0

# clipping color saturation to 0-0.4
sns.heatmap(H,vmin=0,vmax=.4)
plt.show()
```
